2. Ordenar los elementos de una lista según el número de vocales que tiene cada palabra.

```
#Se define una función que cuente las vocales de cada palabra, usaremos el modulo re
def N_vocales(palabra):
  '''
  Función que calcula el número de vocales en una palabra
  input: str
  output: el número de vocales de str
  '''
  try:
    find=re.findall(r'[aeiouAEIOU]', palabra) #se crea una lista con los elementos que coincidan con los caracteres [aeiouAEIOU]
    return len(find)
  except:
    return 'introduce un str'
#Ahora se procede a definir una función que ordene las listas que le demos en función del número de vocales
def Orden_Vocales(lst):
  '''
  Función que ordena una lista de str
  '''
  try:
    # usamos sorted() para ordenar la lista, se colocan como argumentos la lista a ordenar
    # reverse ordena los elemenos de forma desendente
    # key recibe una función con la cual ordenar, si no se especifica entonces ordena de manera alfabética o numérica
    # En este caso le decimos que ordene de acuerdo al número de vocales de cada una de las palabras
    orden=sorted(lst, reverse=True, key=lambda cada_palabra: N_vocales(cada_palabra))
    return orden
  except:
    return 'introduce una lista que contenga puros datos tipo srt'
```

3. Escriba una función que tome dos listas y devuelva True si tienen al menos un miembro común.

```
def Uno_en_comun(lst1,lst2):
  '''
  Función que busca al menos una coincidencia entre elementos de 2 listas
  input: lista1, lista2
  output: srt, bool.
  '''
  control=False #Establecemos una variable que se modifique si encuentra coincidencias
  for i in lst1: #iniciamos bucle tomando cada elemento de la primer lista
    for j in lst2: #comparamos cada elemento de lst1 con todos los de lst2
      if i==j:
        control=True #Si encuentra coincidencia, la variable control cambia a True
        return(control)
      else:
        continue #Si no encuentra coincidencias, la función sigue comparando con los demás elementos de lst1
  if control==False: #Si no hay ninguna coincidencia, se lanza mensaje
    return 'no se encontraron coincidencias'
```

4. Escriba un programa para obtener un string a partir de un string dado, donde todas las apariciones de su primer carácter se han cambiado a '\$', excepto el primer carácter en sí.

Sample String : 'restart'

Expected Result : 'resta$t'

```
def cambio(cadena):
    '''
    Función que toma una cadena, toma su primer caractér y sustituye su aparición por '$', excepto el primero
    input: str
    output: srt modificada
'    '''
    try:
      objetivo=cadena[0] #guardamos el primer caractér de la cadena
      restante=cadena[1:] #Guardamos el resto de la cadena, menos el primer caractér
      #Creamos la nueva cadena pegando el primer caractér con el resto de ella remplazando dicho caractér con '$'
      nueva=cadena[0]+restante.replace(objetivo,'$')
      return nueva
    except:
      return 'introduce un srt'
```

5. Escriba un programa que tome un string y calcula el número de dígitos y letras que contiene.

Ejemplo: Dado el string "hola23", debería devolver:

letras = 4
dígitos = 2

```
#Definimos una cadena de prueba
cadena='21n1231jbs,0.2ihaosan,?281$$,aojoa'
def numeros_y_letras(cadena):
  #Contamos las letras usando la función 'sum'
  #Dentro de la función se itera sobre toda la cadena y busca las letras, sumando 1 cada que encuentra una
  letras=sum(1 for i in cadena if i.isalpha())
  #Contamos los números usando la función 'sum'
  #Dentro de la función se itera sobre toda la cadena y buscan los números, sumando 1 cada que encuentra uno
  numeros=sum(1 for i in cadena if i.isdigit())
  print(f'Hay {numeros} números\nHay {letras} letras')

numeros_y_letras(cadena)
```

7. Escriba un programa para verificar la validez de la contraseña ingresada por los usuarios. Con las siguientes condiciones:

- Al menos 1 letra entre [a-z] y 1 letra entre [A-Z].

- Al menos 1 número entre [0-9].

- Al menos 1 carácter de [$ # @].

- Longitud mínima 6 caracteres.

- Longitud máxima de 16 caracteres.

```
def contraseña(cadena):
  '''
  Función que aprueba o deniega una contraseña, la cual debe tener al menos una letra entre [A-Z], otra entre [a-z],
  un número y un caractér especial (ej. $%),.)
  input: str
  output: str aprobando y denegando la validez de la contraseña
  '''
  #Buscamos las letras mayusculas
  Letras=len(re.findall(r'[A-Z]',cadena))
  #Buscamos las letras minúsculas
  letras=len(re.findall(r'[a-z]',cadena))
  #Buscamos los números
  Numeros=len(re.findall(r'[0-9]',cadena))
  #Buscamos los caractéres especiales
  especiales=len(re.findall(r'[¡!@#$%^&*()_+{}\[\]:;<>,.¿?~\\/\|=]',cadena))
  #Multiplicamos el número de valores encontrados
  cond=Letras*letras*Numeros*especiales
  #Si alguna longitud de las listas tenía 0 valores, entonces cond==0
  #Además pedimos que tenga más de 6 caractéres y menos de 16
  if (cond==0)|(len(cadena)<6)|(len(cadena)>16):
    #Si es inválida, enviamos mensaje de la forma válida de contraseñas
    print('intruduce una contraseña con lo siguiente:\nuna letra entre[a-z]\
    y otra entre [A-Z]\nal menos un número y un caractér especial\nmínimo 6 caracteres y máximo 16')
  else:
    #Caso contrario aceptamos la constraseña como válida
    print('contraseña válida')
```

8. Escribe un programa para contar el número de ocurrencias de cada carácter en una cadena.

Ejemplo: "Mariana"

"M":1

"a":3

"r:1

"i":1

"n":1


```
def conteo_letras(palabra):
  '''
  función que cuenta la ocurrencia de los caractéres dentro de una cadena
  input: str
  output: dict
  '''
  #definimos un diccionario para guardar cada caractér único con su respectivo conteo
  dic={}
  #Interamos sobre todo el str
  for letra in palabra:
    #si el caractér no se encuentra en la str entonces lo guarda y cuenta 1
    if letra not in dic:
      dic[letra]=1
    else:
      #Caso contrario, suma uno más
      dic[letra]+=1
  return dic
```

9. Haz un función para que tome como parámetro un número n y una cadena no vacía, que devuelva una cadena con el n-ésimo carácter eliminado.


```
def n_eliminado(n,cadena):
  '''
  función que elimina el n_esimo caracter de una cadena de caracteres
  input: n=int,cadena=srt
  output: srt
  '''
  try:
    nueva_cadena=cadena[:(n-1)]+cadena[n:] #creamos una nueva cadena que sea el resultado de pegar 2 cadenas
    return nueva_cadena                    #que sean resultado de cortar antes y despues de la aparición del n-ésimo caracter
  except:
    return 'introduce un numero entero n en el rango y una cadena'
```

[Hint](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html)


*Sugerencia: Podemos invertir valores booleanos utilizando el simbolo ~ en Series de Pandas

```
#EJEMPLO
a_serie = pd.Series([True, False, True])
print(~a_serie)
```

```
#Tu código aquí :)

#Buscamos los elementos iguales de 1 a 2
igual_serie1a2=serie_1.isin(serie_2)
print('los iguales en la serie 1 que en la serie 2 son:\n')
for i in range(len(serie_1)):
  if igual_serie1a2[i]==True:
    print(serie_1[i])
  else:
    continue
#buscamos los diferentes de 1 a 2
dif_serie1a2=~igual_serie1a2
print('los diferentes en la serie 1 que en la serie 2 son:\n')
for i in range(len(serie_1)):
  if dif_serie1a2[i]==True:
    print(serie_1[i])
  else:
    continue
#buscamos los iguales de serie 2 a 1
igual_serie2a1=serie_2.isin(serie_1)
print('los iguales en la serie 2 que en la serie 1 son:\n')
for i in range(len(serie_2)):
  if igual_serie2a1[i]==True:
    print(serie_2[i])
  else:
    continue
#buscamos los diferentes de serie 2 a 1
dif_serie2a1=~igual_serie2a1
print('los diferentes en la serie 2 que en la serie 1 son:\n')
for i in range(len(serie_2)):
  if dif_serie2a1[i]==True:
    print(serie_2[i])
  else:
    continue
```

¿Podrías juntar los elementos que NO se comparten entre las series en una misma serie?

```
#Tu código aquí :)

#Sí se puede pues tienen el mismo tamaño y quedaría de la siguiente forma:

#Creamos 2 series con los valores diferentes de cada uno de ellos
serie_aux1=pd.Series([serie_1[i] for i in range(len(serie_1)) if dif_serie1a2[i]==True])
#Despues lo mismo pero los elementos del serie 2 diferentes de la 1
serie_aux2=pd.Series([serie_2[i] for i in range(len(serie_2)) if dif_serie2a1[i]==True])
#finalmente convertimos ambos en un diccionario con zip() y lo convertimos en un series

Elem_dif=pd.Series(dict(zip(serie_aux1,serie_aux2)))
Elem_dif
```
